```python
#Beautiful Soup
from bs4 import BeautifulSoup
import requests
import os
import pandas as pd

#set up
url = 'http://adventuretime.wikia.com/wiki/Episodes'

#Requests Package
at = requests.get(url)
soup = BeautifulSoup(at.text, "lxml")
```

```python
results = soup.find_all(style='text-align: center; background: #f2f2f2')
```

```python
links = []
for r in results:
    link = r.find_all('a')
    try:
        links.append(link[1].get('href'))
    except IndexError:
        pass
```

```python
short_url = 'http://adventuretime.wikia.com'
all_text = []
episode_count = 1

for l in links:
    at = requests.get(short_url + l + '/Transcript') #adds transcript
    print(short_url + l + '/Transcript')
    soup = BeautifulSoup(at.text, "lxml")
    dls = soup.find_all('dl')
    
    #gets the text
    for dl in dls:
        all_text.append([episode_count, dl.get_text()])
        
    episode_count += 1
```

```python
df = pd.DataFrame(all_text)
df.columns = ['episode', 'text']
save_path = os.getcwd() + '\\..\\data\\all_text.csv'
df.to_csv(save_path, index = False)
```
